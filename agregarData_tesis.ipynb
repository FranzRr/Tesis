##Tweepy

```python
from config import *
import tweepy
import datetime
import pandas as pd
import re
```

```python
#Sacando variables del archivo config.py que es donde están las claves de mi cuenta de Developer
#de Twitter.
auth = tweepy.OAuthHandler(TWITTER_CONSUMER_KEY, TWITTER_CONSUMER_SECRET)
auth.set_access_token(TWITTER_ACCESS_TOKEN, TWITTER_ACCESS_TOKEN_SECRET)
api = tweepy.API(auth)
```

```python
#Se usará la fecha actual, y también la fecha hasta el día de ayer
today = datetime.date.today()
date= today - datetime.timedelta(days=1)
today, date
```

```python
all_tweets = []
palabra = "Ousama ranking"
for tweet in tweepy.Cursor(api.search, q=palabra + " until:" + str(today), tweet_mode ="extended", lang="es").items(100):
  all_tweets.append(tweet.full_text)
```

```python
all_tweets[1]
```

```python
df_tweepy = pd.DataFrame()
df_tweepy["tweets"] = all_tweets
```

```python
df_tweepy
```

```python
for i in range(df_tweepy.shape[0]):
  nueva_fila = {"tweet": df_tweepy["tweets"][i], "Sentimiento": ""}
  df_final = df_final.append(nueva_fila, ignore_index=True)
```
